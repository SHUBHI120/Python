```python
#drop command
import numpy as np
dict={"rollno": [1,2,3,4],
    "Name": ["Shubhangi","Tanul","Mallika","Shirya"],
    "Branch": ["CSE","CSE","ME",np.nan],
    "Sem":[4,5,6,4]}
s=pd.DataFrame(dict)

new_s = s.drop(2, axis=0)
print(new_s)
```

```python
#fillna command
import numpy as np
dict={"rollno": [1,2,3,4],
    "Name": ["Shubhangi","Tanul","Mallika","Shirya"],
    "Branch": ["CSE","CSE","ME",np.nan],
    "Sem":[4,5,6,4]}
s=pd.DataFrame(dict)

new_s = s.fillna("CE", axis=0)
print(new_s)
```
