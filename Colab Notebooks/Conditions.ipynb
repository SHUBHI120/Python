```python
Sno1=int(input("enter marks of 1 subject:"))
Sno2=int(input("enter marks of 2 subject:"))
Sno3=int(input("enter marks of 3 subject:"))
Sno4=int(input("enter marks of 4 subject:"))
Sno5=int(input("enter marks of 5 subject:"))
avg=(Sno1+Sno2+Sno3+Sno4+Sno5)/5

if(avg>=90):
   print("Grade A")
elif(avg>=80 and avg<90):
   print("Grade B")
elif(avg>=70 and avg<80):
   print("Grade C")
elif(avg>=60 and avg<70):
   print("Grade D")
```

```python
num = int(input("Enter a number: "))    
fact = 1    
if num < 0:    
   print(" Factorial does not exist for negative numbers")    
elif num == 0:    
   print("The factorial of 0 is 1")    
else:    
   for i in range(1,num + 1):    
       fact = fact*i    
   print("The factorial of",num,"is",fact)  
```
