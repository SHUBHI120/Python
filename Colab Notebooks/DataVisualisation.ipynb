```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x="temperature",data=data)
plt.title("temperature")
plt.show()
```

```python
sns.countplot(x="rainfall",data=data)
plt.title("rainfall")
plt.show()
```

```python
sns.distplot(data.humidity)
plt.title("humidity range")
plt.show()
```

```python
print(data["rainfall"].value_counts())
```

```python
hist=data[["temperature","humidity","rainfall","ph"]].hist(bins=10,figsize=(10,10),color='yellow')
print(hist)
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(x=data['temperature'],y=data['humidity'], hue=data['rainfall'])
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.show()
```

```python
plt.figure(figsize=(5,8))
sns.countplot(x='temperature',data=data)
plt.title("humidity")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(data.corr(),annot=True)
plt.title('correlation between columns')
plt.show()
```
